```python
class Solution:
    def twoSum(self, nums, target):
        pair_idx = {}

        for i, num in enumerate(nums):
            if target - num in pair_idx:
                return [i, pair_idx[target - num]]
            pair_idx[num] = i           
```

```python
def twoSum(nums, target):
 n=len(nums)
 for i in range(0,n-1):
      for j in range(i+1,n):
        if nums[i]+nums[j]==target:
            return [i,j]
print(twoSum([2,7,11,15],9))        
```

```python
def sum(nums,target):
    n=len(nums)
    for i in range(0,n-1):
        for j in range(i,n):
            if nums[i]+nums[j]==target:
                return [nums[i],nums[j]]
print(sum([1,2,3,4],7))            
```

```python
def func(s, left, right):
    if left >= right:
        return True
    
    if s[left] != s[right]:
        return False
    
    return func(s, left + 1, right - 1)

print(func("nitin", 0, len("nitin") - 1))
```

```python
def intToRoman(num):
    values  = [1000, 900, 500, 400, 100, 90, 50, 40,
               10, 9, 5, 4, 1]

    symbols = ["M", "CM", "D", "CD", "C", "XC", "L", "XL",
               "X", "IX", "V", "IV", "I"]

    result = ""

    for i in range(len(values)):
        count = num // values[i]   # how many times it fits
        result += symbols[i] * count
        num = num % values[i]     # reduce number

    return result


# Test
print(intToRoman(1994))
```

```python
grid = [
    [2, 2],
    [2, 2]
]

def equal_partition(grid):
    total = 0

    # total sum
    for row in grid:
        total += sum(row)

    if total % 2 != 0:
        return False

    target = total // 2

    # horizontal
    cur=0
    for row in grid:
     cur+=sum(row)
     if cur==target:
        return True
     col=0
     for col in range(grid[0]):
        cur+=grid[0][col]+grid[1][col]
        if cur==target:
           return True
     return False
print(equal_partition(grid))       
```

```python
haystack = "hello"
needle = "ll"
class Solution:
    def heystack(self, haystack, needle):
       n,m=len(haystack),len(needle)
       for i in range(n-m+1):
           if haystack[i:i+m]==needle:
               return i
       return -1
o=Solution()     
print(o.heystack(haystack,needle))    
```
